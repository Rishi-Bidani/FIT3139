```python
def runge_kutta2(f, x0, y0, b, h, num_iter):
    x = x0
    y = y0

    # alpha * beta = 1/2
    beta = 1 / (2 * b)
    alpha = 1 / (2 * beta)
    # a + b = 1
    a = 1 - b

    y_values = []
    for _ in range(num_iter):
        k1 = f(x, y)
        k2 = f(x + alpha * beta, y + beta * h * k1)
        y_values.append(k2)
        x += h
        y += h * (a*k1 + b*k2)
    
    return y_values
```

```python
f = lambda x, y: x*y
runge_kutta2(f, 0, 0, 5, 1, 100)
```

```python
def RK_next(df, y, t, h, b):
    if b ==0:
        a, alpha, beta = 1, 1, 1
    else:
        a, alpha, beta = 1-b, 1/(2*b), 1/(2*b)
    
    k1 = df(t, y)
    k2 = df(t + (alpha*h), y + beta*h*k1)

    return y + h * (a*k1 + b*k2)
```

```python
def RK2(f, x0, y0, b, h, num_iter):
    t = x0
    y = y0

    # alpha * beta = 1/2
    beta = 1 / (2 * b)
    alpha = 1 / (2 * beta)
    # a + b = 1
    a = 1 - b

    y_values = []
    x_values = []
    z_values = []

    for _ in range(num_iter):
        x = RK_next(f[0], y, t, h, b)
        x_values.append(x)
        y = RK_next(f[1], y, t, h, b)
        y_values.append(y)
        z = RK_next(f[2], z, t, h, b)
        z_values.append(z)
        t += h

    return x_values, y_values, z_values
    
```

```python
sigma = 10
beta = 8/3
rho = 28

dxdt = lambda t, xyz: sigma*(xyz[1] - xyz[0])
dydt = lambda t, xyz: xyz[0] * (rho - xyz[2]) - xyz[1]
dzdt = lambda t, xyz: xyz[0]*xyz[1] - beta*xyz[2]
functions = [dxdt, dydt, dzdt]

x0, y0, z0 = 1,1,1
```
