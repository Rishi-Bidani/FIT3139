```python
import numpy as np
import matplotlib.pyplot as plt
```

# r = 0.5

```python
p0 = 5
p = [p0]
t = [0]
r = 0.5
k = 100
for i in range(100):
    t.append(i)
    next_p = p[-1] + r*p[-1]*(1-(p[-1]/k))
    p.append(next_p)
plt.plot(t, p)
```
