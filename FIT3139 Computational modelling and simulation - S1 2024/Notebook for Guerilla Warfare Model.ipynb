```python
from scipy.integrate import odeint
```

```python
a = 20
b = 5
def guerilla_warfare(P, t=0):
    return np.array([-a*P[1] ,
                     -b*P[0]*P[1] ])
```

```python
t = linspace(0, 2,  500)              # time
P0 = array([10, 10])                     # initials conditions: 10 rabbits and 5 foxes
P, info = odeint(guerilla_warfare, P0, t, full_output=True)
home, enemy = P.T
plot(t, home, label='home')
plot(t, enemy, label='enemy')
plt.grid()
plt.legend()
```
