# Lets try it out

```python
schelling_1 = schelling.Schelling(width=50,height=50,
                                      empty_ratio=0.3, 
                                      similarity_threshold=0.55,
                                      n_iterations=1000)
```

```python
schelling_1.plot()
```

```python
unsatisfied_agents = schelling_1.go()
```

```python
plt.plot(unsatisfied_agents)
plt.xlabel('Time')
plt.ylabel('# unsatisfied agents')
plt.grid()
```
